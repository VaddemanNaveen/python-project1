```python
## Display the percentage distribution of the data in each year using the pie chart.
```

```python
# Calculate the total honey production for each year
yearly_production = data1.groupby('year').totalprod.sum().reset_index()

# Calculate the percentage of honey production for each year
yearly_production['percent'] = (yearly_production['totalprod'] / yearly_production['totalprod'].sum()) * 100

# Create a pie chart of the percentage distribution of honey production by year
plt.pie(yearly_production['percent'], labels=yearly_production['year'], autopct='%1.1f%%')
plt.title('Percentage Distribution of Honey Production by Year')
plt.show()
```
